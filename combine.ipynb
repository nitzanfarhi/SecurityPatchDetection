```python
import os
import pandas as pd
```

```python
for idx,file in enumerate(os.listdir("D:/hidden_cve/graphql2/")):
    if idx==166:
        continue
    commitsA = pd.read_csv("D:/hidden_cve/graphql2/"+file)
    try:
        commitsB = pd.read_csv("D:/hidden_cve/graphql/"+file.split("_")[1])
    except FileNotFoundError:
        continue
    res = pd.concat([commitsB.drop(["additions","deletions","commit_date"],axis=1),commitsA],axis=1)
    res.to_csv("D:/hidden_cve/graphql/"+file.split("_")[1])
    #print(idx,file,commitsA.shape,commitsB.shape,res.shape)
```
